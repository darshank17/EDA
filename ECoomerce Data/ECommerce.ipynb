# E Commerce Dataset EDA , Model Training




**Problem Statement**:

Predict the purchase amount of customer against the products which will help them to create personalised offer for customers against different products



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# import Train Dataset
df_train = pd.read_csv('/content/train.csv.zip')
df_train.head()
```

```python
# import test data
df_test = pd.read_csv('/content/test.csv.zip')
df_test.head()
```

```python
# Merge Both datset
df = pd.concat([df_train,df_test])
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
# If you want to Deletet a Column ( axis = 0 -> row & axis = 1 -> Column)
df.drop(['User_ID'],axis=1,inplace = True)
```

```python
# Preproccesing the categorial data i.r changing string -> nums

# Now i want to change the Gender Categorial data into nums
# We will use map function

df['Gender'] = df['Gender'].map({'F':0 ,'M': 1})  # F -> 0 & M -> 1
```

```python
# Handel Age ( Its in range )

df['Age'].unique()
```

```python
df['Age'] = df['Age'].map({'0-17': 1, '18-25':2 ,'26-35': 3, '36-45':4,'46-50':5,'51-55' : 6,'55+':7  })
```

```python
df.head()
```

```python
df_city = pd.get_dummies(df['City_Category'], drop_first=True,dtype = int)
df_city.head()
```

```python
df=pd.concat([df,df_city], axis = 1)
df.head()
```

```python
df.drop('City_Category', axis =1 , inplace = True)
```

```python
# Missing value
df.isnull().sum()
```

```python
## Focus on replacing missing value
df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].value_counts()
```

```python
df['Product_Category_2'].mode()[0]
```
