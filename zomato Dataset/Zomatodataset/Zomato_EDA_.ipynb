```python
df_country = pd.read_excel('/content/sample_data/Country-Code.xlsx')
df_country
```

```python
# We have mergerd the two datasets on the columns country code and in the left df
final_df = pd.merge(df,df_country,on ="Country Code" , how = "left" )
```

Exploring The Dataset


```python
# Records according to country
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
## Pie Chart - Top 3 countries using zomato
plt.pie(country_values[:3] , labels = country_names[:3] )
```

Obsv : Majority Sales are in India


```python
final_df.columns
```

```python
## Renaming The Last column from 0 -> Rating Count

ratings = final_df.groupby(['Aggregate rating', "Rating color", 'Rating text']).size().reset_index().rename(columns={0: 'Rating Counts'})
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot( x = "Aggregate rating" , y = "Rating Counts" ,hue = "Rating color", data = ratings, palette = ['blue', 'red','Orange','Yellow','Green','Green' ] )

```
